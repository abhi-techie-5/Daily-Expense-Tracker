```python
# Daily Expense Tracker


import numpy as np

def Daily_expense(days):
  num=[]
  for i in range(days):
    x=int(input(f"Enter the daily expenses(in Rupee): "))
    num.append(x)
  return np.array(num)

def Expenses_stats(arr):
  Total_Expense=np.sum(arr) # get Total Expense
  Avg_Expense=np.mean(arr) # get average Expense
  High_ExpenseDay=np.argmax(arr)+1  # get highest Expense day

  return Total_Expense,Avg_Expense,High_ExpenseDay


def display(Total_Expense,Avg_Expense,High_ExpenseDay):
  print(f"Total expense is given as: {Total_Expense} Rupee  ")
  print(f"Average expense is given as: {Avg_Expense} Rupee ")
  print(f"Highest expense was on day {High_ExpenseDay} ")


  # optional Warning
  limit=1000

  if Avg_Expense >limit:
    print("Warning !: Average expense exceeds the limit.")

  else:
    print("Average expense is within the limit.")
days =int(input("Enter number of days: "))
arr=Daily_expense(days)
Total_Expense,Avg_Expense,High_ExpenseDay=Expenses_stats(arr)
display(Total_Expense,Avg_Expense,High_ExpenseDay)
```
